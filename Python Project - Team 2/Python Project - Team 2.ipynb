## Project Goal
The aim of this project is to conduct an exploratory analysis - using python - over World Bank data in the region: Middle East and Russia.
 
## Introducing the Region
The Middle East and Russia region is made up of 13 countries that are located in Eastern Europe, the middle east and Eurasia. Countries: Ukraine, Moldova, Romania, Bulgaria, Belarus, Georgia, Azerbaijan, Armenia, Iran, Turkmenistan, Uzbekistan, Afghanistan and Pakistan. These countries are highly dependent on natural resources such as petroleum, natural gas and coal (CIA, 2020).
 
Russia is not geographically close to the region but has important historic influence on some of the countries. This influence goes from an economic and political perspective to even a religious point of view. Economically, countries such as Belarus, Moldova, Ukraine and Armenia rely heavily on Russian energy exports (CIA, 2020). Additionally, some of the countries have maintained Soviet-style command economies with subsidies and strict controls on production prices and access to foreign currency. About half of the countries are Orthodox while the other half are Muslims (CIA, 2020).

## Middle East & Russia vs The World
In order to compare the Middle East and Russia with the rest of the world, we decided to highlight topics related to healthcare, carbon dioxide emissions, gender equality and internet access. 
 
### HIV treatment (Appendix: Step 7.2.1)
Of all the people who have HIV in the Middle East, only 5% (median) of them are receiving antiretroviral therapy, in contrast the world median is 18%. “Life expectancy study has shown that HIV-positive people who start antiretroviral therapy (ART) promptly and have good access to medical care live as long as their HIV-negative peers” (Aidsmap, 2020). This issue needs to be addressed by the governments of middle eastern countries.
 

### Incidence of tuberculosis (Appendix: Step 7.2.2)
The amount of people that develop tuberculosis (TB) per 1 thousand people in the Middle East (107 people) is almost two times the world equivalent (57 people). That being said, “Over 95% of TB deaths occur in low- and middle-income countries” (WHO, 2020), and most of the countries in our region fall in this category. 
 
### Carbon Dioxide (Co2) Emissions (Appendix: Step 7.2.3)
The region’s dependency on fossil fuels results in CO2 emissions being higher than the world standard, at 5 metric tons per capita as opposed to the world median at 3 metric tons per capita. As it is being done in countries such as the United Kingdom, governments in the region could encourage businesses to control their emissions by giving a tax break or a subsidy.
 
### Access to the Internet (Appendix: Step 7.2.4)
Information and communication technologies promote opportunities for economic growth, improved health and education (Worldbank, 2020). Access to the internet in the Middle East and Russia region (12%) is almost half of what the rest of the world has (22%). Upon further analysis, “it is becoming increasingly common for governments in the MENA region to block content that originates from or is affiliated with rival states in the region” (Berkman Klein Center, 2019).
 
 
### Gender Equality (Appendix: Step 7.2.5)
According to the World Bank (2020), gender inequalities remain pervasive in many dimensions of life worldwide and women are vastly underrepresented in decision-making positions in government. Without representation at this level, it is difficult for women to influence policies. The world median shows that only 15% of seats in National Parliaments are held by women. In comparison, the median for the region is 13%. The lower number held by the Middle East could be a consequence of a culture and religion that believes in the superiority of men against women. 

### Understand if the data can be trusted
The data was collected in 2011 and after almost 10 years of political and economic events in the region, it can be said that most of the countries remain in the same position they were one decade ago. In other words, the fact that events such as the US pull out of the Iran’s nuclear deal and external regional wars, has kept the whole region in the same economic and political situation. Therefore, the data used for this analysis is still relevant. However, we need to have in mind that the countries might have developed strategies along the years to contain or improve some of the issues identified and aspects such as internet access have had a major shift in a global scale over the last decade. 

 
## Conclusion
The analysis of the Middle East and Russia has provided key insights on the socio-economic issues present within the region. The region mainly consists of underdeveloped countries. Afghanistan and Pakistan were outliers on a significant portion of variables analyzed, showing that they are struggling more with some issues than the rest of the region.

The comparison to the rest of the world highlights the differences that exist in a region with a majority of countries being third world countries versus the world average. There are many areas of improvement within the region that governments need to dedicate time and resources on. Moreover, the heavy dependency on Russia and China for the majority of their imports should be reconsidered as it makes these countries even more vulnerable.
